## Tests B1 and E1 Generation: BERT Masked-token filling

```python
from transformers import pipeline, AutoTokenizer, AutoModel
```

```python
model_name = "bert-base-uncased"
```

```python
### code inspired from Hugging Face, https://huggingface.co/docs/transformers/main_classes/pipelines, last access 20 March
fill_mask= pipeline("fill-mask", model= model_name)
```

```python
def template_filling (pipeline, template):
    """
    Fill in a template with masked tokens using the predefined pipeline.

    paramater:
    -pipeline: The hugging face pipeline
    -template: String with ONE [MASK]

    prints: the filled-in sentences
    """
   
    results= pipeline(template, top_k= 50)
    for case in results:
        print(case["sequence"])
```

### B1:

B1 evaluates whether an SRL model predicts semantic roles based on the predicate argument structure rather than lexical cues (e.g., word classes and frequency). 
In this notion, B1 reflects predicate substitution with the fixed contexts, as in the sentence pair below:
- The house **costs** *me* <ARG3> \$100,000.
- The house **takes** *me* <ARG2> \$100,000

In both contexts, there is a buyer, 'me', buying 'the house' with `\$100,000`. However, with the predicate change from `costs' to `takes', a model is expected to label semantic functions of `me' differently (i.e., `a buyer` or `thing required from`), reflecting the DIR test type.

This notebook thus aims to generate use cases for Test B1. Here templates are crafted (e.g., `He [MASK] *me* <ARG3> \$100,000.`) and BERT is used to suggest the predicates to fill in. The generation is subject to manual inspections.

```python
#template 1
t1= 'The house [MASK] me $100,000 .'
uc1= template_filling (fill_mask, t1)
```

```python
#template 2
t2= 'Selling the house [MASK] me $100,000 .'
uc2= template_filling (fill_mask, t2)
```

```python
#template 3
t3= 'He [MASK] me $ 100,000  for the house .'
uc3= template_filling (fill_mask, t3)
```

```python
#template 4
t4= 'She [MASK] me the house for $100,000 .'
uc4= template_filling (fill_mask, t4)
```

```python
#template 5
t5= 'The house price [MASK] to $100,000 .'
uc5= template_filling (fill_mask, t5)
```
